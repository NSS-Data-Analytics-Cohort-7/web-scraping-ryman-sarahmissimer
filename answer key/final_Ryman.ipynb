```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS
import re
import matplotlib as plt
import seaborn as sns
from lxml import html
```

```python
page = 1
i = 1
artists = []
date = []
time = []
openers = []

while page != 6:
    url = f"https://ryman.com/events/list/?tribe_event_display=list&tribe_paged={page}"
    print(url)
    response = requests.get(url)
    soup = BS(response.text)
    for h2 in soup.find_all(attrs={'class': 'tribe-event-url'}):
        artists.append(h2.get_text(strip=True))
    for p in soup.find_all('time'):
        p = p.get_text(strip=True)
        p = str(p).split(' at ')
        date.append(p[0])
        time.append(p[1])
    for event in soup.find_all('div', {'class': 'tribe-beside-image'}):
        try:
            opener = event.find('span', {'class': 'opener'})
            if opener:
        # Sometimes there is a 2nd Show Added!, which is also considered to be an opener
        # Grab the last one
                opener = event.find_all('span', {'class': 'opener'})[-1]
                if re.match('with (special guest)?', opener.text):
                        openers.append(re.search('with (special guests? )?(.+)', opener.text)[2])
                else:
                    openers.append("none")
            else:
                openers.append("none")
        except:
            opener.append("none")
    page = page + 1
    
df1 = {'Artist':artists,'Date':date, 'Time': time, 'Opener': openers}
df1 = pd.DataFrame(df1)
df1
```
